Question 1: What is a random variable in probability theory?
-  A random variable in probability theory is a real-valued function that assigns a numerical value to each possible outcome of a random experiment. It is denoted by a capital letter (e.g.,(X)) and its specific value is determined by chance after the experiment is conducted.
    - Key Concepts
    - Function: A random variable is a function where the input is an outcome from the sample space (the set of all possible results of a random experiment) and the output is a real number.
    - Randomness: The "random" aspect comes from the fact that the underlying experiment's outcome is uncertain, so the numerical value the variable takes on is also uncertain before the experiment.
    - Probability Distribution: Each possible value (or range of values) of the random variable has an associated probability, described by a probability distribution.
      - Types of Random Variables:
        - Random variables are primarily classified into two types:
        - Discrete Random Variable: A variable that can take on a finite or countably infinite set of distinct values (typically whole numbers), such as the number of heads in a coin toss or the number of calls a support center receives in an hour.
        - Continuous Random Variable: A variable that can take any value within a given continuous range or interval, such as the height of a person or the time it takes for a process to complete.


Question 2: What are the types of random variables?
- In statistics, random variables are primarily classified into two main types: discrete random variables and continuous random variables. A third, less common type, is the mixed random variable.
  - 1. Discrete Random Variables
      - A discrete random variable is one that can take on a finite or countably infinite number of distinct values. These variables usually arise from counting something.
        - Characteristics:
            - Values are specific and separate (often integers).
            - There are "gaps" between the possible values.
            - The probability distribution is described by a Probability Mass Function (PMF), which assigns a probability to each individual value.
        - Examples:
            - The number of heads in three coin tosses (possible values: 0, 1, 2, or 3).
            - The number of cars sold by a dealership in a month.
            - The number of defective light bulbs in a box of ten.
            - The outcome of rolling a die (possible values: 1, 2, 3, 4, 5, or 6).
  - 2. Continuous Random Variables
      - A continuous random variable can take any value within a given range or interval. These variables usually arise from measuring something.
        - Characteristics:
            - Values form an infinite and uncountable set within a specific range.
            - The probability of the variable taking on any single, exact value is zero.
            - The probability distribution is described by a Probability Density Function (PDF), and probabilities are assigned to intervals (represented by the area under the curve of the PDF).
        - Examples:
            - The height or weight of a person.
            - The time it takes to run a mile.
            - The amount of rainfall in a city over a year.
            - The temperature of a room.
  - 3. Mixed Random Variables
      - A mixed random variable has a cumulative distribution function that is a combination (or "mixture") of a discrete random variable's function and a continuous random variable's function. It has both discrete and continuous components.
        - Characteristics:
            - Can take specific, countable values with positive probability, as well as any value within continuous intervals.
        - Example:
            - A system where a person's wait time in a queue is 0 minutes if the server is free (a discrete outcome), but a continuous variable (e.g., anywhere between 0 and 10 minutes) if the server is busy.
   

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.
- Mean, median, and mode are the three primary measures of central tendency, which are single values used to represent the center or typical value of a dataset. They are important because they provide a concise summary of the data and help in understanding its distribution.
  - Definitions
    - Mean: Also known as the arithmetic average, the mean is calculated by summing all the values in a dataset and then dividing by the number of values. It is a widely used measure but can be significantly affected by outliers (extreme values).
        - Use Case: Best for data that is symmetrically distributed without extreme outliers, such as the average height of people in a room.
    - Median: The median is the middle value of a dataset when the data is ordered from least to greatest. If there is an odd number of values, the median is the single middle number. If there is an even number of values, the median is the average of the two middle numbers.
        - Use Case: Best for data that may have outliers or is skewed, as the median is not affected by extreme values.
    - Mode: The mode is the value that appears most frequently in a dataset. A dataset can have one mode (unimodal), more than one mode (multimodal), or no mode if all values appear only once.
        - Use Case: Primarily used for categorical or nominal data, such as the most popular color of cars in a parking lot.
    - Importance of Measures of Central Tendency
      - These measures are important for several reasons:
        - Summarizing Data: They provide a single, representative value that condenses large amounts of data into an understandable format, making it easier to communicate and interpret findings.
        - Facilitating Comparisons: They allow for quick comparisons between different datasets or groups. For example, comparing the average test scores (mean) of two different classes.
        - Identifying Typical Values: They help in identifying what a "normal" or "common" data point looks like within a distribution, offering insight into the center of the data.
        - Understanding Data Distribution: By comparing the mean, median, and mode, one can infer the shape or skewness of the data's distribution. For instance, if the mean is significantly different from the median, the data is likely skewed.
        - Informing Decision Making: In various fields like business, science, and economics, these measures help in making informed decisions and predictions. For example, a retailer might use the mode to decide which product size to stock the most.
          

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?
- Skewness and kurtosis are statistical measures that describe the shape of a data distribution.
- Skewness
  - Skewness measures the asymmetry of a data distribution. A perfectly symmetrical distribution, like a normal (bell-shaped) curve, has a skewness of zero.
      - Zero Skewness: The distribution is a mirror image on both sides of the center point (mean = median = mode).
      - Positive Skewness (Right-Skewed): The tail of the distribution is longer on the right side. Most of the data is concentrated on the left side of the graph, and a few high values pull the mean higher than the median.
      - Negative Skewness (Left-Skewed): The tail of the distribution is longer on the left side. Most of the data is concentrated on the right side of the graph, and a few low values pull the mean lower than the median.
- Kurtosis: Kurtosis measures the "tailedness" of a data distribution, which describes the shape of the peak and the thickness of its tails relative to a normal distribution. It gives insight into the probability of extreme values (outliers).
  - Mesokurtic: The distribution has a kurtosis similar to a normal distribution (excess kurtosis of 0, or an absolute kurtosis of 3).
  - Leptokurtic: The distribution has higher kurtosis (positive excess kurtosis) with a sharp peak and heavy, long tails, indicating a higher likelihood of outliers or extreme events.
  - Platykurtic: The distribution has lower kurtosis (negative excess kurtosis) with a flatter peak and light, short tails, indicating a lower probability of extreme events.
- What a Positive Skew Implies
  - A positive skew (or right skew) implies several things about the data:
    - Tail Direction: The "long tail" of the distribution extends towards the right-hand side (the positive direction on a number line).
    - Data Concentration: The majority of the data points are clustered on the lower end of the scale, around the left side of the graph.
    - Central Tendency Relationship: The mean is typically greater than the median, which is greater than the mode (Mean > Median > Mode). This is because the extreme high values in the right tail pull the mean in their direction, while the median (the middle value) is less affected.
    - Real-World Example: Income distribution is a classic example of a positively skewed dataset; most people earn a moderate income, but a few individuals earn extremely high incomes, which stretches the right tail.



```python
# Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers. numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
# Ans
import collections

def compute_statistics(numbers):
    # Step 1: Calculate the Mean
    mean = sum(numbers) / len(numbers)

    # Step 2: Calculate the Median
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 0:
        # Even number of elements
        median = (sorted_numbers[n // 2 - 1] + sorted_numbers[n // 2]) / 2
    else:
        # Odd number of elements
        median = sorted_numbers[n // 2]

    # Step 3: Calculate the Mode
    counts = collections.Counter(numbers)
    max_count = max(counts.values())
    mode = [key for key, value in counts.items() if value == max_count]

    return mean, median, mode

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Call the function and print the results
mean, median, mode = compute_statistics(numbers)

print(f"The list of numbers is: {numbers}")
print(f"The mean is: {mean}")
print(f"The median is: {median}")
print(f"The mode(s) is/are: {mode}")
```

```python
# Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60]
# Ans.
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to NumPy arrays
array_x = np.array(list_x)
array_y = np.array(list_y)

# Compute covariance
# np.cov returns a covariance matrix. The covariance between x and y is at index [0, 1] or [1, 0].
covariance_matrix = np.cov(array_x, array_y)
covariance = covariance_matrix[0, 1]

# Compute correlation coefficient
correlation_coefficient = np.corrcoef(array_x, array_y)[0, 1]

print(f"Covariance between list_x and list_y: {covariance}")
print(f"Correlation coefficient between list_x and list_y: {correlation_coefficient}")
```

```python
# Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
# Ans.
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.title('Boxplot of Numeric Data')
plt.ylabel('Value')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Calculate quartiles and IQR to identify outliers
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
print(f"Identified Outliers: {outliers}")
```

```python
# Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales. ● Explain how you would use covariance and correlation to explore this relationship. ● Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000]
# Ans.
#As a data analyst, you can use covariance and correlation to determine if a linear relationship exists between advertising spend and daily sales.
#How to use Covariance and Correlation to Explore the Relationship
# 1. Covariance
# Explanation: Covariance measures the direction of the linear relationship between two variables. A positive covariance indicates that as advertising spend increases, daily sales tend to increase. A negative covariance indicates that as advertising spend increases, daily sales tend to decrease. A covariance near zero suggests no strong linear relationship.
# Limitation: Covariance's value is dependent on the units of the variables (e.g., dollars spent and dollars earned), making it difficult to compare the strength of the relationship across different datasets.
# 2. Correlation
# Explanation: Correlation (specifically, the Pearson correlation coefficient) standardizes the measure of the linear relationship by dividing the covariance by the product of the standard deviations of both variables. The result is a value between -1 and +1.
# Interpretation:
# A value of +1 means a perfect positive linear relationship (sales increase proportionally with spend).
# A value of -1 means a perfect negative linear relationship (sales decrease proportionally with spend).
# A value of 0 means no linear relationship.
# Advantage: Correlation provides a clear, unitless measure of the strength and direction of the relationship, allowing for easy interpretation and comparison.
# Summary of Approach:
# First, calculate the covariance to determine the direction of the relationship. Then, calculate the correlation to determine the strength of that relationship. A high positive correlation would suggest that increasing advertising spend is associated with an increase in daily sales, which is valuable information for the marketing team.

import numpy as np

# Data provided
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert lists to NumPy arrays for calculation
spend_array = np.array(advertising_spend)
sales_array = np.array(daily_sales)

# Compute the correlation coefficient
# np.corrcoef returns a correlation matrix.
# We extract the value for the correlation between the two variables (e.g., [0, 1] or [1, 0]).
correlation_matrix = np.corrcoef(spend_array, sales_array)
correlation_coefficient = correlation_matrix[0, 1]

print(f"The advertising spend data: {advertising_spend}")
print(f"The daily sales data: {daily_sales}")
print(f"The correlation coefficient between advertising spend and daily sales is: {correlation_coefficient:.4f}")
```

```python
# Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product. ● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use. ● Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
# Ans
# To understand the distribution of customer satisfaction survey data (on a scale of 1-10) before launching a new product, you would use a combination of central tendency, variability, and positional statistics, along with a visualization that shows the frequency of each score.
# Summary Statistics and Visualizations to Use
# Here are the recommended statistics and visualizations:
# Summary Statistics
# Statistic 	Purpose
# Mean	Measures the average satisfaction score, giving a central value for the data [1].
# Median	Represents the middle score when all scores are ordered; it is less affected by extreme values than the mean and is useful for ordinal data like a 1-10 scale [1].
# Mode	Identifies the most frequent score(s) in the dataset, showing the typical response from customers [1].
# Standard Deviation	Measures the dispersion or spread of the data points around the mean. A small standard deviation means scores are clustered close to the mean, while a large one means they are more spread out [1].
# Range	The difference between the maximum and minimum scores, indicating the overall span of satisfaction levels [1].
# Visualizations
# Histogram: The most effective visualization for understanding the distribution of a single numerical variable. It will display the frequency (count) of each score (1 through 10) in a bar chart format, clearly showing where most scores fall and if the distribution is skewed [1].
# Box Plot (Box-and-Whisker Plot): This plot provides a visual summary of the median, quartiles (25th and 75th percentiles), and potential outliers. It is excellent for quickly understanding the data's spread and identifying the central 50% of responses [1].

import matplotlib.pyplot as plt

# The provided survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Define the bins for the histogram
# Since the data is discrete (integers 1-10), we set the bins
# to cover each integer value clearly.
# We set bins from 4 to 10 with a width of 1 to ensure each score has its own bar.
# Alternatively, we can use a range of bins:
min_score = min(survey_scores)
max_score = max(survey_scores)
# Bins should cover all possible values; we use a range that goes slightly beyond
bins = range(min_score, max_score + 2) # Creates bins: 4 to 11 (covers scores 4-10)

# Create the histogram
plt.figure(figsize=(8, 5)) # Optional: Adjust figure size for better readability
plt.hist(survey_scores, bins=bins, edgecolor='black', alpha=0.7)

# Add titles and labels
plt.title('Distribution of Customer Satisfaction Survey Scores (1-10 Scale)')
plt.xlabel('Survey Score')
plt.ylabel('Frequency')

# Set the x-axis ticks to show only integer values corresponding to scores
plt.xticks(range(min_score, max_score + 1))

# Add a grid for better readability (optional)
plt.grid(axis='y', alpha=0.5)

# Display the plot
plt.show()
```
